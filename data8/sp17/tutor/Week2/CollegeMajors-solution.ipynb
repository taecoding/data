It's good to think about what we can use visualization for. Here's an example of something that might not be terribly insightful: plot the employement rate vs unemployment rate for all majors. 

Create and add a new column called employment rate to bothsorted. (This is employed/(unemployed + employed)). Assign this new table to the same name as before, bothsorted.

```python
bothsorted = bothsorted.with_column("Employment_rate", bothsorted.column("Employed")/(bothsorted.column("Unemployed") + bothsorted.column("Employed")))
```

```python
bothsorted.scatter("Employment_rate", "Unemployment_rate")
```
